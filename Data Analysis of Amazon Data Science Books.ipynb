```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("final_book_dataset_kaggle.csv")
```

```python
df.tail()
```

```python
df.nunique()
```

```python
df1 = df.dropna(subset = ['author'])
```

```python
df1['price'] = df['price'].replace(np.nan, df['price'].mean())
```

```python
df1['pages'] = df['pages'].replace(np.nan, df['pages'].mean())
```

```python
df1['avg_reviews'] = df['avg_reviews'].replace(np.nan, df['avg_reviews'].mean())
```

```python
df1['n_reviews'] = df['n_reviews'].replace(np.nan, df['n_reviews'].mean())
```

```python
df1['language'] = df['language'].replace(np.nan,'English')
```
